# Unique identifier for each individual (customer).

- Own_property- Binary feature indicating whether the individual owns a property (0 for no, 1 for yes).

- Work_phone- Binary feature indicating whether the individual has a work phone (0 for no, 1 for yes).

- Phone- Binary feature indicating whether the individual has a phone (0 for no, 1 for yes).

- Email- Binary feature indicating whether the individual has provided an email address (0 for no, 1 for yes).

- Unemployed- Binary feature indicating whether the individual is unemployed (0 for no, 1 for yes).

- Num_children- Number of children the individual has.

- Num_family- Total number of family members.

```python
df=df.drop(['ID',"Phone","Email",'Work_phone','Num_family','Unemployed'],axis=1)
#dropping the uneccessary columns 
```

```python
gender_counts = df['Gender'].value_counts()
total_individuals = len(df)
male_count = gender_counts[0]
female_count = gender_counts[1]
male_proportion = male_count / total_individuals
female_proportion = female_count / total_individuals
print("the female proportion is :",female_proportion*100,'%')
print("the male proportion is :",male_proportion*100,'%')
```

```python
gender = df['Gender'].value_counts()
plt.figure(figsize=(7, 6))
ax = gender.plot(kind='bar', rot=0, color="c")
ax.set_title("Bar Graph of Gender")
ax.set_xlabel('Gender')
ax.set_ylabel('Number of People')
ax.set_xticklabels(('Male', 'Female'))

for rect in ax.patches:
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2
    space = 1
    label = format(y_value)
    ax.annotate(label, (x_value, y_value), xytext=(0, space), textcoords="offset points", ha='center', va='bottom')    
plt.show()
```

```python
#checking for gender and the eligibility of credit card
gender_eligible=df[df['Target']==1]['Gender'].value_counts()
gender_not_eligible=df[df['Target']==0]['Gender'].value_counts()
print("gender and Credit Card Eligibility:")
print(f"Eligible for a Credit Card:\n{gender_eligible}")
print(f"\nNot Eligible for a Credit Card:\n{gender_not_eligible}")
```

```python
#classifying on the basis of ownership of car 
car=df["Own_car"].value_counts()
total_car=len(df['Own_car'])
car_own=car[1]
no_car=car[0]
car_pro=round((car_own/total_car)*100,2)
no_carpro=round((no_car/total_car)*100,2)
print("proportion of people having car:",car_pro,"%")
print("proportion of people not having car:",no_carpro,"%")
```

```python
colors = ['#4287f5', '#f54242']
plt.figure(figsize=(8, 6))
bars = plt.bar(['Owns a Car', 'Does not Own a Car'], [car_own, no_car], color=colors)

plt.show()
```

```python
#classification on the basis of the possession of automobile and the target variable
car_1=df[df['Target']==1]['Own_car'].value_counts()
car2=df[df['Target']==0]['Own_car'].value_counts()
print("classification on the basis of the possession of automobile and the target variable")
print(" eligible for credit card:",car_1)
print("not eleigible for credit card ",car2)
```

```python
#Classification on the basis of ownership of property
prop=df['Own_property'].value_counts()
total_prop=len(df['Own_property'])
own_prop=prop[1]
not_own_prop=prop[0]
own_ratio=round((own_prop/total_prop)*100,2)
not_own_ratio=round((not_own_prop/total_prop)*100,2)
print("The percentage of people owning property:",own_ratio,"%")
print("The percenyage of people not owning property:",not_own_ratio,"%")
```

```python
plt.figure(figsize=(7,7))
colors = ['#4287f5', '#f54242']
bar_prop=plt.bar(["OWN a property","Does not own property"],[own_prop,not_own_prop],color=['red','yellow'])
plt.show()

```

```python
prop_own=df[df['Target']==1]['Own_property'].value_counts()
prop_not_own=df[df['Target']==0]['Own_property'].value_counts()
print(" eligible for credit card:",prop_own)
print("not eleigible for credit card ",prop_not_own)
```

```python
num_children_counts = df['Num_children'].value_counts().sort_index()
num_children_counts.index
color = '#FF5733'
plt.figure(figsize=(10, 6))
bars = plt.bar(num_children_counts.index, num_children_counts.values, color=color, edgecolor='black')
plt.grid(color='gray', linestyle='dashed', linewidth=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tick_params(axis='x', colors='gray')
plt.tick_params(axis='y', colors='gray')
plt.title('Distribution of Number of Children for Individuals', fontsize=16, fontweight='bold', color='black')
plt.xlabel('Number of Children', fontsize=14, fontweight='bold', color='gray')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='gray')
plt.tight_layout()
plt.show()
```

```python
income_eligible = df[df['Target'] == 1]['Total_income']
income_not_eligible = df[df['Target'] == 0]['Total_income']

print("\nIncome and Credit Card Eligibility:")
print(f"Eligible for a Credit Card:\n{income_eligible}")
print(f"Not Eligible for a Credit Card:\n{income_not_eligible}")
```

```python
occupation_types = df['Occupation_type'].unique()

#individuals based on their occupation type
occupation_counts = df['Occupation_type'].value_counts()
```

```python
colors = sns.color_palette('pastel')
plt.figure(figsize=(10, 6))
bars = plt.bar(occupation_counts.index, occupation_counts.values, color=colors)
for bar in bars:
    height = bar.get_height()
    plt.annotate('{}'.format(height),
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3), 
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=12, fontweight='bold')
plt.grid(color='gray', linestyle='dashed', linewidth=0.5)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tick_params(axis='x', colors='gray')
plt.tick_params(axis='y', colors='gray')
plt.title('Individuals based on Occupation Type', fontsize=16, fontweight='bold', color='black')
plt.xlabel('Occupation Type', fontsize=14, fontweight='bold', color='gray')
plt.ylabel('Count', fontsize=14, fontweight='bold', color='gray')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
